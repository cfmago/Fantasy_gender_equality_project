```python
# GoT plot

n_groups = 5

males = (79, 73, 52, 87, 76)
females = (21, 26, 48, 13, 24)

abs_males = [1469, 22, 149, 408, 1061]
abs_females = [395, 8, 140, 62, 333]

fig, ax = plt.subplots(figsize=(12,8))

index = np.arange(n_groups)
bar_width = 0.35

rects1 = plt.bar(index, males, bar_width, label='Men', alpha= 0.8)
rects2 = plt.bar(index + bar_width, females, bar_width, label='Women', alpha= 0.8)

#plt.xlabel('Word')
#plt.ylabel('Frequency')
plt.title('GOT - Distribution by gender', fontsize = 16)
plt.xticks(index + bar_width/2, ('All_characters (1864)', 'POV_chapter (30)', 'Married (289)', 'Dead (470)', 'Alive (1394)'), fontsize = 12)
plt.legend(loc = 'best', fontsize = 12)

plt.rcParams['font.size'] = 12

for rect in rects1:
    height = rect.get_height()
    ax.text(rect.get_x() + rect.get_width()/2., 1*height, '%d' % float(height) + "%", ha='center', va='bottom')
    
for rect in rects2:
    height = rect.get_height()
    ax.text(rect.get_x() + rect.get_width()/2., 1*height, '%d' % float(height) + "%", ha='center', va='bottom')
    
plt.savefig("GoT_characters.jpg")

plt.tight_layout()
plt.show()
```
