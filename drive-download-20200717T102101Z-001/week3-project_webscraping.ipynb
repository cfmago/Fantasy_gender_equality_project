```

url_list = 'https://www.goodreads.com/list/show/143005.Absolute_favorite_books_I_ve_read_part_2'
response = requests.get(url_list)
soup = BeautifulSoup(response.content, features = 'lxml')
#get ratings
book = soup.find_all('span', attrs = {'class': 'minirating'})
books = book[:5]
book_ratings = [ book.get_text(strip = True) for book in books ]

ratings = []

for rating in book_ratings:
  ratings.append({'rating': rating})

ratings
```

```
ratings = pd.DataFrame(ratings)
ratings
```

```
booktitle = [ {'booktitle' : 'A Dance with Dragons'},
            {'booktitle' : 'A Feast for Crows'},
            {'booktitle' : 'A Storm of Swords'},
            {'booktitle' : 'A Clash of Kings'},
            {'booktitle' : 'A Game of Thrones'}]
```

```
booktitle = pd.DataFrame(booktitle)
booktitle
```

```
books_ratings_got = booktitle.join(ratings)
books_ratings_got
```

# From list page: LOTR

```
url_list = 'https://www.goodreads.com/search?q=tolkien&qid='
response = requests.get(url_list)
soup = BeautifulSoup(response.content, features = 'lxml')
#get ratings
book = soup.find_all('span', attrs = {'class': 'minirating'})
books = book[:4]
book_ratings = [ book.get_text(strip = True) for book in books ]

ratings = []

for rating in book_ratings:
  ratings.append({'rating': rating})

ratings
```

```
ratings = pd.DataFrame(ratings)
ratings
```

```
booktitle = [ {'booktitle' : 'The Hobbit, or There and Back Again'},
            {'booktitle' : 'The Fellowship of the Ring'},
            {'booktitle' : 'The Two Towers'},
            {'booktitle' : 'The Return of the King'}]
```

```
booktitle = pd.DataFrame(booktitle)
booktitle
```

```
books_ratings_lotr = booktitle.join(ratings)
books_ratings_lotr
```

```
dfs = [books_ratings_got, books_ratings_lotr]

all_ratings = pd.concat(dfs)
all_ratings
```

```
all_ratings = all_ratings.reset_index()
all_ratings
```

```
all_ratings = all_ratings.drop("index", axis=1)
all_ratings
```

```
# pokemon["Combo Type"] = np.where(pokemon["Type 2"].isnull(), pokemon["Type 1"], pokemon["Type 1"] + "-" + pokemon["Type 2"])
import numpy as np
all_ratings['author'] = ''
all_ratings
```
