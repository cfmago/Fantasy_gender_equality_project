# WEB SCRAPING

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'https://www.goodreads.com/search?q=game+of+thrones+ice+%26+fire&qid=ARlWXLRyib'
response = requests.get(url)
response
```

```python
soup = BeautifulSoup(response.content, features = 'lxml')
soup
```

```python
book_GOT = soup.find_all('span', attrs = {'class': 'minirating'})
book_GOT[0].get_text(strip = True)
```
